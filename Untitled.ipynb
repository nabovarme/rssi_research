```python
info = pd.read_csv("info.csv")

known_ssid = pd.read_csv("ssid.csv")
wifi_scan = pd.read_csv("wifi_scan_view.csv", dtype={'id':int, 'serial':str, 'bssid':str, 'rssi':int, 'channel':str, 'unix_time':int})

wifi_scan['rssi_normalized'] = -1 * ( wifi_scan.rssi - 14 )

known_networks = wifi_scan[wifi_scan.ssid.isin(known_ssid.ssid)]
mesh_networks = wifi_scan[wifi_scan.ssid.str.startswith("mesh-")]
all_networks = pd.concat([known_networks, mesh_networks])

print("len raw:", len(wifi_scan), "len filtered:", len(all_networks), "of known:", len(known_networks), "mesh:", len(mesh_networks))
all_networks.head()
```

```python
G=nx.Graph()
hotspot_nodes = []
mesh_nodes = []

for serial in wifi_scan.serial.unique():
    key = f'mesh-{serial}'
    node = G.add_node(key, node_type='mesh')
    mesh_nodes.append(key)
    
for ssid in known_ssid.ssid.unique():
    key = ssid
    node = G.add_node(key, node_type='hotspot')
    hotspot_nodes.append(key)


for index, row in wifi_scan.iterrows():
    G.add_weighted_edges_from([(f"mesh-{row['serial']}", row['ssid'], row['rssi_normalized'])])

#fig, ax = plt.subplots(1, 1, figsize=(8, 6))
#nx.draw_circular(G, ax=ax, node_size=10, width=0.1)
```

```python
best_paths = []

for mesh_node in tqdm(mesh_nodes):
    candidate_paths = []
    for hotspot_node in hotspot_nodes:
        try:
            path = nx.dijkstra_path(G,mesh_node, hotspot_node)
            if len(path) < 8:
                weights = [G[path[i-1]][path[i]]['weight'] for i in range(1,len(path))]
                candidate_paths.append(dict(
                    path=path,
                    weights=weights,
                    summed_weight=sum(weights),
                    summed_weight_mean=statistics.mean(weights),
                    links=len(path)
                ))
        except KeyboardInterrupt:
            raise
        except nx.NetworkXNoPath as e:
            pass
        except Exception as e:
            print("unknown error", e, 'from', mesh_node)
            raise
    if not candidate_paths:
        print("no candidates for ", mesh_node)
        continue

    best_strength_path = sorted(candidate_paths, key=lambda x: x['summed_weight_mean'])[:1]
    
    best_paths.extend(best_strength_path)
    
```

```python
import json
print(json.dumps(best_paths))
```

```python
tree_graph = nx.DiGraph()

seen = set()


for path in best_paths:
    nodes = path['path']
    
    for index in range(1, len(nodes)):
        node_from = nodes[index-1]
        node_to = nodes[index]
        
        if node_from not in seen:
            node = G.nodes[node_from]
            seen.add(node_from)
            tree_graph.add_node(node_from, node_type=node['node_type'])
            
        if node_to not in seen:
            node = G.nodes[node_to]
            seen.add(node_to)
            tree_graph.add_node(node_to, node_type=node['node_type'])
        try:
            tree_graph[node_to][node_from]
        except:
            tree_graph.add_edges_from([(node_from, node_to)])

```

```python
#print_json

def lookup(serial):
    try:
        return info[info['serial']==serial.replace('mesh-', '')]['info'].values[0] + f" ({serial})"
    except:
        return serial
    
def create_trees(prev_node=None,current_node=None):
    print("evaluating", current_node, flush=True)
    prefix = ""
    if current_node is None:
        incomming_nodes = [str(node) for node in tree_graph.nodes() if tree_graph.out_degree(node) == 0]
        prefix = ''
    else:
        incomming_nodes = [str(x[0]) for x in tree_graph.in_edges(current_node)]

    if not incomming_nodes:
        return ''
    
    returnable = {}
    for incomming_node in incomming_nodes:
        rssi_prefix = ""
        if current_node:
            normalized_rssi = G[incomming_node][current_node]['weight']

            rssi = (-1 * normalized_rssi) + 14

            rssi_prefix = f"rssi: {rssi} name: "
            
        returnable[prefix+rssi_prefix+lookup(incomming_node)] = create_trees(current_node,incomming_node)
    return returnable
trees = create_trees()
import json
print(json.dumps(trees, indent=4, ensure_ascii=False))
```
